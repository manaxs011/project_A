```python
#GET DUMMY -to convert categorical variable into dummy or indicator variable
dummy=pd.get_dummies(df[['Fuel_Type','Transmission','Seller_Type']],drop_first=True)
dummy
```

```python
#concatenation for data manipulation
df1=pd.concat([df,dummy],axis=1)
df1
```

```python
#we are droping unwanted columns
df2=df1.drop(['Car_Name','Fuel_Type','Seller_Type','Transmission'],axis=1)
df2
```

```python
# seperating input and output
x=df2.drop(['Selling_Price'],axis=1)
x
```

```python
y=df2['Selling_Price']
y
```

```python
sns.regplot(x=df['Present_Price'],y=y,color='g')
```

```python
#seperating training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
#preprocessing
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
#model building
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#compare actual value and predicted value by creating new dataframe
df3=pd.DataFrame({'actual_value':y_test,'prediced_value':y_pred})
df3
```

```python
print('slop is',model.coef_)
print('constant is',model.intercept_)
```

```python
#Perform evaluation 
#mean absolute error
from sklearn.metrics import mean_absolute_error
print('mean absoulte error is',mean_absolute_error(y_test,y_pred))
```

```python
#mean absolute percentage error
from sklearn.metrics import mean_absolute_percentage_error
print('mean absolute percentage error is',mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error
print('mean squared error is',mean_squared_error(y_test,y_pred))
```

```python
#root mean squared_error
from sklearn.metrics import mean_squared_error
k=mean_squared_error(y_test,y_pred)
c=np.sqrt(k)
print('root mean squared error is',c)
```
