```python
#fill missing values
df['Weight']=df['Weight'].fillna(df['Weight'].mean())
df['Duration']=df['Duration'].fillna(df['Duration'].mean())
df['Body_Temp']=df['Body_Temp'].fillna(df['Body_Temp'].mean())
df.isna().sum()
```

```python
df1=df.drop(['User_ID'],axis=1)
df1.head()
```

```python
#Converting gender column to numerical values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Gender']=le.fit_transform(df['Gender'])
df1.head()
```

```python
# seperating input and output
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
x.tail()
```

```python
#seperating training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

**Prediction on Test Data Using DecisionTree Algorithm**

```python
#using DecisionTree algorithm
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
#compare actual value and predicted value by creating new dataframe
df2=pd.DataFrame({'actual_value':y_test,'prediced_value':y_pred})
df2
```

```python
#Perform evaluation 
#mean absolute error
from sklearn.metrics import mean_absolute_error
print('mean absoulte error is',mean_absolute_error(y_test,y_pred))
```

```python
#mean absolute percentage error
from sklearn.metrics import mean_absolute_percentage_error
print('mean absolute percentage error is',mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error
print('mean squared error is',mean_squared_error(y_test,y_pred))
```

```python
#root mean squared_error
from sklearn.metrics import mean_squared_error
k=mean_squared_error(y_test,y_pred)
c=np.sqrt(k)
print('root mean squared error is',c)
```
